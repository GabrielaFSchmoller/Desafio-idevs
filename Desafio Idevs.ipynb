```python
def divisores(n):
    qtd_div=0
    for i in range (1,n+1):
        if(n%i)==0:
            qtd_div=qtd_div+1
    return qtd_div

for j in range(1,10000):
    if divisores(j)==divisores(j+1):
        print(j)
```

```python
def primo(n):
    if divisores(n)==2:
        return True
    return False 
```

```python
num=int(input('Digite o número inteiro positivo:'))
i=0
l_primos = []
while i<num:
    i=i+1
    if primo(i) == True:
        l_primos.append(i)
    
```

```python
print(l_primos)
```
